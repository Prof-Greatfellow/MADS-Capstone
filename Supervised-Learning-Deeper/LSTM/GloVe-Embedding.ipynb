```python
import os
import pickle
import numpy as np
```

```python
os.chdir('C:\glove')
```

```python
with open('glove.twitter.27B.200d.txt', encoding='UTF-8') as f:
    words = f.readlines()
```

```python
words[0]
```

```python
word2id = dict()
embeddings = []
```

```python
for i, w in enumerate(words):
    w_list = w.split(' ')
    word2id.update({w_list[0]: i})
    embeddings.append([float(j) for j in w_list[1:]])
```

```python
embeddings[0]
```

```python
embeddings = np.array(embeddings)
os.chdir('C:\\Users\\close\\Documents\\GitHub\\MADS-Capstone\\assets\\glove')
pickle.dump(word2id, open('glove_word2id', 'wb'))
np.save('glove_embeddings', embeddings)
```
