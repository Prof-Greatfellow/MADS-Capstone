```python
p.set_options(p.OPT.URL, p.OPT.MENTION, p.OPT.RESERVED, p.OPT.EMOJI, p.OPT.SMILEY, p.OPT.NUMBER)
df['tweet_text_cleaned'] = df['tweet_text'].apply(p.clean)
```

```python
df['hashtags'] = df['tweet_text_cleaned'].str.lower().str.findall(r"#(\w+)")
```

```python
p.set_options(p.OPT.URL, p.OPT.MENTION, p.OPT.RESERVED, p.OPT.EMOJI, p.OPT.SMILEY, p.OPT.NUMBER, p.OPT.HASHTAG)
df['tweet_text_cleaned'] = df['tweet_text_cleaned'].apply(p.clean)
```

```python
df
```

```python
from gensim.parsing.preprocessing import preprocess_string, strip_tags, strip_punctuation, strip_multiple_whitespaces, strip_numeric
```

https://radimrehurek.com/gensim/parsing/preprocessing.html

```python
CUSTOM_FILTERS = [lambda x: x.lower(), strip_tags, strip_punctuation, strip_multiple_whitespaces, strip_numeric]
df['tweet_text_cleaned'] = df['tweet_text_cleaned'].apply(lambda x:' '.join(preprocess_string(x, CUSTOM_FILTERS)))
```

```python
df['processed_text_length'] = df['tweet_text_cleaned'].str.len()
```

```python
unique_labels = df['class_label'].unique()
dict_label = dict(zip(unique_labels,np.arange(len(unique_labels))))
```

```python
dict_label
```

```python
df['class_label_id'] = df['class_label'].map(dict_label)
```

```python
df = df[~df['tweet_text_cleaned'].isna()]
```

```python
df.to_csv('assets/df_cleaned.csv', index=False)
```
